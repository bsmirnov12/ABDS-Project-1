```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import scipy.stats as stats
df = pd.read_csv("data/incomevsvotes.csv")
df.head(20)
```

```python
df2=df[["Federal Electoral Districts_x", "Income per Resident", "Party Won"]]
df2
```

```python
df3 = df2.dropna()
```

```python
df2.boxplot("Income per Resident", by="Party Won", figsize=(20, 10))
```

```python
group1 = df2[df2["Party Won"] == "Bloc Québécois"]["Income per Resident"]
group2 = df2[df2["Party Won"] == "Conservative"]["Income per Resident"]
group3 = df2[df2["Party Won"] == "Green Party"]["Income per Resident"]
group4 = df2[df2["Party Won"] == "Liberal"]["Income per Resident"]
group5 = df2[df2["Party Won"] == "NDP"]["Income per Resident"]
```

```python
stats.f_oneway(group1, group2, group3, group4, group5)
```
